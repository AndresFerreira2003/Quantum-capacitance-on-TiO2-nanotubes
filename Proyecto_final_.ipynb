### 3. Diagramas de Nyquist y bode##

```python
def fit_circle_least_squares(x, y):
    # Ajuste de circunferencia: (x - xc)^2 + (y - yc)^2 = r^2
    def residuals(c):
        xc, yc, r = c
        return np.sqrt((x - xc)**2 + (y - yc)**2) - r

    # Estimaciones iniciales
    x_m = np.mean(x)
    y_m = np.mean(y)
    r0 = np.mean(np.sqrt((x - x_m)**2 + (y - y_m)**2))
    res = least_squares(residuals, x0=[x_m, y_m, r0])
    xc, yc, r = res.x
    return xc, yc, r
```

```python
def cole_cole(xi, xo, tau, alpha, f):
    """
    Cole-Cole model for dielectric relaxation.

    Parameters:
    xo : float
        Low-frequency limit of the complex permittivity.
    xi : float
        High-frequency limit of the complex permittivity.
    tau : float
        Relaxation time.
    alpha : float
        Distribution parameter (0 < alpha < 1).
    f : array-like
        Frequency values at which to evaluate the model.

    Returns:
    array-like
        Complex permittivity values at the given frequencies.
    """
    w = 2 * np.pi * f
    return xi + (xo - xi) / (1 + (1j * w * tau)**(1-alpha))
```

```python
data_file = ['https://github.com/AndresFerreira2003/Quantum-capacitance-on-TiO2-nanotubes/raw/main/EIS_Nyquist_Bode%20TiO2%2013%20may%202025_placa%201.xlsx', 'https://github.com/AndresFerreira2003/Quantum-capacitance-on-TiO2-nanotubes/raw/main/EIS_Nyquist_Bode%20TiO2%2013%20may%202025_placa%202.xlsx']
data1 = pd.read_excel(data_file[0])
data2 = pd.read_excel(data_file[1])
```

```python
p = {
    "Z'":"Z' (Ω)",
    "Z''":"-Z'' (Ω)",
    "C'":"C' (F)",
    "C''":"C'' (F)",
    "f":"Frequency (Hz)",
    "Z":"Z (Ω)",
    "Ph":"-Phase (°)",
}
```

```python
C1_real = data1[p["C'"]].values
C1_imag = data1[p["C''"]].values

C2_real = data2[p["C'"]].values
C2_imag = data2[p["C''"]].values

# Ajustar circunferencia
xc1, yc1, r1 = fit_circle_least_squares(C1_real, C1_imag)
xc2, yc2, r2 = fit_circle_least_squares(C2_real, C2_imag)
```

```python
t = np.linspace(0, 2 * np.pi, 100)
x1 = xc1 + r1 * np.cos(np.linspace(0, t, 100))
y1 = yc1 + r1 * np.sin(np.linspace(0, t, 100))

xo1 = xc1 + np.sqrt(r1**2 - (yc1 - 0)**2)
xi1 = xc1 - np.sqrt(r1**2 - (yc1 - 0)**2)
phi1 = np.arcsin(-yc1 / r1)
alpha1 = phi1*2/np.pi

t = np.linspace(0, 2 * np.pi, 100)
x2 = xc2 + r2 * np.cos(np.linspace(0, t, 100))
y2 = yc2 + r2 * np.sin(np.linspace(0, t, 100))

xo2 = xc2 + np.sqrt(r2**2 - (yc2 - 0)**2)
xi2 = xc2 - np.sqrt(r2**2 - (yc2 - 0)**2)
phi2 = np.arcsin(-yc2 / r2)
alpha2 = phi2*2/np.pi

```

```python
idx_max_C1_imag = data1[p["C''"]].idxmax()
max_C1_imag = data1[p["C''"]].iloc[idx_max_C1_imag]
freq_at_max_C1_imag = data1[p["f"]].iloc[idx_max_C1_imag]
print(f"Máximo C'' (F): {max_C1_imag}")
print(f"Frecuencia correspondiente (Hz): {freq_at_max_C1_imag}")

tau1 = 1 / (2 * np.pi * freq_at_max_C1_imag)
print(f"Constante de tiempo τ (s): {tau1}")

f1 = np.logspace(1, 6, 1000)  # Frecuencias de 1 Hz a 1 MHz
C_fit01 = cole_cole(xi1, xo1, tau1, alpha1, f1)

Rq1 = tau1 / (xo1 - xi1)

h = 6.62607015E-34
qe = 1.602176634E-19
R0 = h / (qe**2)
print(f"Resistencia Rq (Ω): {Rq1}")
print(f"Capacitancia Cq (uF): {1e6 * (xo1 - xi1)}")
print(f"N = int(R0 / Rq): {int(R0 / Rq1)}")


idx_max_C2_imag = data2[p["C''"]].idxmax()
max_C2_imag = data2[p["C''"]].iloc[idx_max_C2_imag]
freq_at_max_C2_imag = data2[p["f"]].iloc[idx_max_C2_imag]
print(f"Máximo C'' (F): {max_C2_imag}")
print(f"Frecuencia correspondiente (Hz): {freq_at_max_C2_imag}")

tau2 = 1 / (2 * np.pi * freq_at_max_C2_imag)
print(f"Constante de tiempo τ (s): {tau2}")

f2 = np.logspace(-1, 6, 1000)  # Frecuencias de 1 Hz a 1 MHz
C_fit02 = cole_cole(xi2, xo2, tau2, alpha2, f2)

Rq2 = tau2 / (xo2 - xi2)

h = 6.62607015E-34
qe = 1.602176634E-19
R0 = h / (qe**2)
print(f"Resistencia Rq (Ω): {Rq2}")
print(f"Capacitancia Cq (uF): {1e6 * (xo2 - xi2)}")
print(f"N = int(R0 / Rq): {int(R0 / Rq2)}")
```
